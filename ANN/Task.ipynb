```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report
import tensorflow
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense , Activation
from tensorflow.keras.activations import relu , softmax
import sys
import warnings
warnings.filterwarnings("ignore")
import time
```

```python
class task :
  def __init__(self) :
    try :
      data = pd.DataFrame(np.random.random((1000 , 24)))
      target = []
      for i in range(1000) :
        if i < 400 :
          target.append(0)
        elif i < 700 :
          target.append(1)
        else :
          target.append(2)
      data["target"] = target

      self.data = data
      self.X = self.data.iloc[: , :-1]
      self.y = self.data.iloc[: , -1]
      self.X_train , self.X_test , self.y_train , self.y_test = train_test_split(self.X , self.y , test_size = 0.2 , random_state=42)
      self.y_perfect = tensorflow.keras.utils.to_categorical(self.y_train , num_classes= len(self.data["target"].unique()))

      # Architecture
    except Exception as e :
      er_type ,er_msg , line_no = sys.exc_info()
      print(f'{er_type} >> {er_msg} >> {line_no.tb_lineno}')

  def training(self) :
    try :
      self.model = Sequential()
      self.model.add(Dense(units = 160 , activation= relu , kernel_initializer="he_uniform" , input_dim = self.X_train.shape[1]))
      self.model.add(Dense(units = 80 , activation= relu , kernel_initializer="he_uniform"))
      self.model.add(Dense(units = 40 , activation= relu , kernel_initializer="he_uniform"))
      self.model.add(Dense(units = 20 , activation= relu , kernel_initializer="he_uniform"))
      self.model.add(Dense(units = 10 , activation= relu , kernel_initializer="he_uniform"))

      self.model.add(Dense(units = len(self.data["target"].unique()) , activation= softmax , kernel_initializer="glorot_uniform"))

      self.model.compile(optimizer = "adam" , metrics = ["Acc"] , loss = "categorical_crossentropy")

      # training
      before_time = time.time()
      self.model.fit(self.X_train , self.y_perfect , batch_size = 50 , validation_split = 0.2 , epochs = 50)
      after_time = time.time()
      print(f'Time Taken for Model Training {after_time - before_time}')

      self.y_test_pred = self.model.predict(self.X_test)
      p = []
      for i in self.y_test_pred :
        p.append(np.argmax(i))

      print(accuracy_score(self.y_test , p))
      print(confusion_matrix(self.y_test , p))
      print(classification_report(self.y_test , p))

    except Exception as e :
      er_type ,er_msg , line_no = sys.exc_info()
      print(f'{er_type} >> {er_msg} >> {line_no.tb_lineno}')

if __name__ == "__main__" :
  try :
    obj = task()
    obj.training()

  except Exception as e :
    er_type ,er_msg , line_no = sys.exc_info()
    print(f'{er_type} >> {er_msg} >> {line_no.tb_lineno}')
```
